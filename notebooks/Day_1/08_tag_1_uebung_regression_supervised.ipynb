# Übung: Überwachtes Lernen - Regression mit linearen Modellen

In dieser Aufgabe bearbeiten Sie ein vollständiges, aber bewusst einfach gehaltenes Regressionsproblem. Ziel ist nicht nur ein Modell zu trainieren, sondern den gesamten Weg von Rohdaten bis Bewertung nachvollziehbar zu dokumentieren.

**Datensatz:** `sklearn.datasets.load_diabetes`

**Erlaubte Bibliotheken:** Nutzen Sie nur Pakete, die in `requirements.txt` enthalten sind und in den bisherigen Kursnotebooks verwendet wurden: `numpy`, `pandas`, `matplotlib` und `scikit-learn`. Verwenden Sie keine zusätzlichen Visualisierungs- oder ML-Bibliotheken.

Der Datensatz enthält Messwerte von Patientinnen und Patienten. Die Zielgröße ist ein kontinuierlicher Krankheitsprogressionswert. Sie sollen daraus ein überwachtes Regressionsmodell entwickeln.

**Wichtig:** Nutzen Sie nur überwachtes Lernen. Clustering, PCA als unüberwachtes Lernziel oder Reinforcement Learning sind in dieser Übung nicht vorgesehen.
