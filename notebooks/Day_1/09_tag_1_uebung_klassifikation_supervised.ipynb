# Übung: Überwachtes Lernen - Klassifikation mit einem linearen Modell

In dieser Aufgabe bearbeiten Sie ein vollständiges binäres Klassifikationsproblem. Sie verwenden ein lineares Klassifikationsmodell, konkret logistische Regression. Die logistische Regression nutzt eine lineare Entscheidungsfunktion und ist daher ein passender nächster Schritt nach linearer Regression.

**Datensatz:** `sklearn.datasets.load_breast_cancer`

**Erlaubte Bibliotheken:** Nutzen Sie nur Pakete, die in `requirements.txt` enthalten sind und in den bisherigen Kursnotebooks verwendet wurden: `numpy`, `pandas`, `matplotlib` und `scikit-learn`. Verwenden Sie keine zusätzlichen Visualisierungs- oder ML-Bibliotheken.

Der Datensatz enthält medizinische Messwerte zu Zellproben. Die Zielgröße ist eine von zwei Klassen. Die Aufgabe ist methodisch gedacht: Trainieren Sie ein überwachtes Modell, bewerten Sie es sauber und reflektieren Sie Grenzen und Risiken.

**Wichtig:** Nutzen Sie nur überwachtes Lernen. Keine Clusteranalyse, keine unüberwachten Labels, kein Reinforcement Learning.
