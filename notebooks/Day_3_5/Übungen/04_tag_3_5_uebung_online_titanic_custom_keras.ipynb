# Übung: Online-Dataset im Kaggle-Einsteigerstil mit Custom Keras

In dieser Übung arbeiten Sie mit einem frei online verfügbaren Titanic-CSV. Titanic ist ein klassisches Einsteigerproblem auf Kaggle: Aus Passagierdaten soll vorhergesagt werden, ob eine Person überlebt hat.

Damit die Übung ohne Kaggle-Login funktioniert, wird ein öffentlich erreichbarer CSV-Mirror genutzt. In einer echten Kaggle-Umgebung können Sie stattdessen die Kaggle-Dateien verwenden.

**Online-CSV:** `https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv`

**Aufgabe:** Binäre Klassifikation mit tabellarischen Daten, Keras-Preprocessing und mindestens einem eigenen Keras-Baustein.

## Lernziele

- einen realen Online-Datensatz laden
- Datenqualität, fehlende Werte und Leakage-Risiken prüfen
- numerische und kategorische Features vorbereiten
- ein Keras-Modell für Tabular-Daten bauen
- eigene Keras-Metrik, eigene Loss-Funktion oder eigene Layer verwenden
- Modellvarianten vergleichen und sauber validieren
- Fehler fachlich interpretieren
