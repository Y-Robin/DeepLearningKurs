# Übung 1: Bilddaten, Filter und Pooling

Diese Übung ist für einen längeren Nachmittag gedacht. Arbeiten Sie nicht nur Code ab, sondern dokumentieren Sie Shapes, Beobachtungen und Fehlerquellen.


## Aufgabe 1 - Eigenes Paint-Bild erzeugen

Erzeugen Sie mit PIL ein RGB-Bild, speichern Sie es als PNG und JPEG und geben Sie einen Pixelmatrix-Ausschnitt aus.


## Aufgabe 2 - Bild als Tensor

Wandeln Sie Ihr Bild in `N,C,H,W` um. Notieren Sie Shape, dtype und Wertebereich.


## Aufgabe 3 - Filterbank

Erstellen Sie mindestens vier Filter: vertikale Kante, horizontale Kante, Weichzeichner, Schärfen. Visualisieren Sie Filtermatrix und Feature Map.


## Aufgabe 4 - Pooling-Vergleich

Vergleichen Sie Max, Average, Min und Global Average Pooling. Testen Sie zusätzlich Padding und eine kleine Verschiebung des Bildinhalts.


## Aufgabe 5 - Kurze Auswertung

Welche Filter reagieren auf welche Bildstellen? Welche Pooling-Methode verliert die meisten Ortsinformationen? Wo entsteht Positionsinvarianz?
