# Übung 4: Modell speichern, laden und reproduzierbar machen

Ziel: Ein trainiertes Modell so speichern, dass es später eindeutig wiederhergestellt werden kann.


## Aufgabe 1 - Custom-Modell bauen

Definieren Sie ein CNN mit mindestens einem eigenen `nn.Module`, zum Beispiel einer eigenen Aktivierung oder einem eigenen Klassifikationskopf.


## Aufgabe 2 - Training und bester Checkpoint

Trainieren Sie das Modell. Speichern Sie nur den besten Checkpoint nach Validierungsaccuracy.


## Aufgabe 3 - Laden und prüfen

Laden Sie das Modell in eine neue Instanz. Prüfen Sie, dass die Vorhersagen auf einem Batch identisch sind.


## Aufgabe 4 - Metadaten

Speichern Sie zusätzlich Klassenliste, Modellkonfiguration, Random Seed, beste Accuracy und Trainingsdatum.


## Aufgabe 5 - Kurze Dokumentation

Schreiben Sie eine Markdown-Zelle, wie eine andere Person Ihr Modell laden und verwenden kann.
