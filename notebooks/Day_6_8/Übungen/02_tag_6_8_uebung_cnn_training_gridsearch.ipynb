# Übung 2: CNN trainieren und Hyperparameter testen

Ziel: Ein eigenes kleines CNN trainieren, systematisch verändern und die Effekte messen. Diese Übung sollte inklusive Auswertung einen Nachmittag füllen.


## Aufgabe 1 - Datensatz wählen

Nutzen Sie `load_digits()` für schnellen Einstieg oder CIFAR-10 für realistischere Bilder. Erstellen Sie Train/Validation/Test-Loader.


## Aufgabe 2 - Baseline-CNN direkt definieren

Definieren Sie ein CNN ohne Architektur-Loop, nur mit `nn.Sequential`. Trainieren Sie mindestens 10 Epochen.


## Aufgabe 3 - Grid Search

Testen Sie mindestens acht Kombinationen aus Filteranzahl, Kernelgröße, Pooling-Methode, Dropout und Lernrate. Speichern Sie Parameteranzahl, Trainingszeit und Validierungsaccuracy in einer Tabelle.


## Aufgabe 4 - Bestes Modell analysieren

Visualisieren Sie Filter und Feature Maps des besten Modells. Zeigen Sie mindestens acht Fehlklassifikationen.


## Aufgabe 5 - Entscheidung begründen

Welches Modell würden Sie abgeben? Begründen Sie anhand von Validierungsleistung, Laufzeit, Stabilität und Modellgröße.
