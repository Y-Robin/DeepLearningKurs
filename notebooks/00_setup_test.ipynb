# Setup-Test für den Deep-Learning-Kurs

Dieses Notebook prüft, ob die wichtigsten Bibliotheken für den Kurs installiert sind und ob TensorFlow/Keras ein kleines neuronales Netz trainieren kann.



## 1. Python- und Paketversionen


```python
import importlib
import platform

packages = [
    'numpy',
    'pandas',
    'scipy',
    'matplotlib',
    'sklearn',
    'tensorflow',
    'tensorboard',
    'transformers',
    'datasets',
    'huggingface_hub',
]

print(f'Python: {platform.python_version()}')
for package in packages:
    module = importlib.import_module(package)
    version = getattr(module, '__version__', 'Version nicht verfügbar')
    print(f'{package:16s} {version}')
```

## 2. TensorFlow-Geräte prüfen


```python
import tensorflow as tf

print('TensorFlow:', tf.__version__)
print('Verfügbare Geräte:')
for device in tf.config.list_physical_devices():
    print('-', device.device_type, device.name)
```

## 3. Kleines Keras-Modell trainieren


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow import keras

rng = np.random.default_rng(42)
x = rng.normal(size=(800, 2)).astype('float32')
y = ((x[:, 0] ** 2 + x[:, 1] ** 2) > 1.0).astype('float32')

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42, stratify=y
)

model = keras.Sequential([
    keras.layers.Input(shape=(2,)),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(8, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, validation_split=0.2, epochs=8, batch_size=32, verbose=0)

pred = (model.predict(x_test, verbose=0).ravel() >= 0.5).astype('float32')
print(f'Test Accuracy: {accuracy_score(y_test, pred):.3f}')

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoche')
plt.ylabel('Loss')
plt.title('Setup-Test: Trainingsverlauf')
plt.legend()
plt.show()
```

## Ergebnis

Wenn alle Zellen ohne Importfehler ausgeführt werden und das kleine Modell trainiert, ist das Basis-Setup für den Kurs einsatzbereit.
